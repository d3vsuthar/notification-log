```python
df['timestamp']= pd.to_datetime(df['postTime'],unit='ms')
df['timestamp'] = df['timestamp'].apply(lambda x: x + dt.timedelta(hours=4))
df['timestamp_str'] = df['timestamp'].apply(lambda x: dt.datetime.strftime(x, "%a %H:%M %p"))
```

```python
df = df[df['group'] != 'ranker_group']
```

```python
df = df[~df['title'].isna()]

apps_of_interest = [
    'in.amazon.mShop.android.shopping',
    'com.namshi.android',
    'com.apparel.app6thstreet',
    'com.sivvi.app',
    'com.sssports.sssports',
    'com.asos.app',
    'com.noon.buyerapp',
    'com.altayer.ounassapplication',
]

df = df[df['packageName'].isin(apps_of_interest)]
```

```python
df['push_body'] = np.where(df['text'].isna(),df['textBig'],df['text'])
```

```python
chart = alt.Chart(df).mark_circle().encode(
    x='timestamp',
    y='appName',
    tooltip=['title', 'push_body', 'timestamp_str']
).properties(
    height=100,
    width=500
)

rules = alt.Chart(pd.DataFrame({
  'Date': [
           '2021-11-21 00:00:00',
           '2021-11-22 00:00:00',
           '2021-11-23 00:00:00',
           '2021-11-24 00:00:00',      
           ],
})).mark_rule().encode(
  x='Date:T'
)
```
